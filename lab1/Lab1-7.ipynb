````python
# print(data['Chronic Diseases'].head(30))

q_75 = data['BMI'].quantile(0.75)
q_03 = data['BMI'].quantile(0.3)
q_06 = data['BMI'].quantile(0.6)

g_75 = data[(data['BMI'] > q_75)]
g_03_06 = data[(data['BMI'] >= q_03) & (data['BMI'] <= q_06)]

g_75_dis = g_75[~g_75['Chronic Diseases'].isnull()]
g_03_06_dis = g_03_06[~g_03_06['Chronic Diseases'].isnull()]


print(f"75%:\n{len(g_75_dis) / len(g_75)}")
print(f"30-60%:\n{len(g_03_06_dis) / len(g_03_06)}")
````

7. Красиво визуализируйте как связаны между собой уровень дохода, когнитивные способности и уровень стресса*

\* Задание на доп. балл

````python
plt.figure(figsize=(12, 8))
sns.scatterplot(x='Cognitive Function', y='Stress Levels', hue='Income Level', data=data, palette='Set1')
plt.title('Связь когнитивной функции с уровнем стресса в зависимости от уровня дохода')
plt.xlabel('Когнитивная функция')
plt.ylabel('Уровень стресса')
plt.legend(title='Уровень дохода')
plt.show()
````

````python

mean_data = data.groupby('Income Level').agg({'Cognitive Function': ['mean', 'sem'], 'Stress Levels': ['mean', 'sem']}).reset_index()

mean_data.columns = ['Income Level', 'Cognitive Function Mean', 'Cognitive Function SEM', 'Stress Levels Mean', 'Stress Levels SEM']
plt.figure(figsize=(12, 8))
palette = sns.color_palette('Set1', n_colors=len(mean_data['Income Level'].unique()))

for i, income_level in enumerate(mean_data['Income Level'].unique()):
    subset = mean_data[mean_data['Income Level'] == income_level]
    plt.errorbar(subset['Cognitive Function Mean'], 
                    subset['Stress Levels Mean'],
                    xerr=subset['Cognitive Function SEM'],
                    yerr=subset['Stress Levels SEM'],
                    fmt='x',
                    color=palette[~i],
                    label=income_level,
                    alpha=0.8)
plt.title('Средние значения когнитивной функции и уровня стресса по уровням дохода')
plt.xlabel('Средняя когнитивная функция')
plt.ylabel('Средний уровень стресса')
plt.legend(title='Уровень дохода')
plt.grid(True)
plt.show()
````
